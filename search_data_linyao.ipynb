```python
import scapy
import requests
import bs4
import re
import json
```

```python
#添加请求头
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36 QIHU 360SE'
}
```

```python
url='http://www.mafengwo.cn/search/q.php?q=%E6%BD%AE%E5%B7%9E&seid=2D00A480-A09F-4E06-A4A9-C8D098554049'

response = requests.get(url,headers=headers)

response.encoding='utf-8'
```

```python
soup = bs4.BeautifulSoup(response.text,"html.parser")

journals = soup.find_all('a', class_='_j_search_link recommend-content-wraper')
```

```python
print(journals)
```

```python
print(journals[0].attrs['href'])
```

```python
urls_all = []
for item in journals:
    urls_all.append(item.attrs['href'])
    
urls_all
```

```python
pattern = re.compile(r'\s|\n|<.*?>',re.S)
content_all = []
content_str = ''
for i in range(3): #因为只有前三个文章是同一个html格式
    
    data = requests.get(urls_all[i],headers=headers).text
    soup_cur = bs4.BeautifulSoup(data,'html.parser')
    tempall = soup_cur.find_all('div',class_='p-section')
    
    for temp in tempall:
        content = temp.get_text()
        content_real = pattern.sub('',content)
        print(content_real)
        content_all.append(content_real)
        content_str+=content_real
    

    
    #print(content_all)
    print('--------------------------------\n\n')
    
    

    
    
```

```python
print(content_str)
```

```python
pattern1 = re.compile(r'[|]|,|，|。|’|‘|;|:|；|：|、|【|】|（|）|',re.S)

content_string_cut = pattern1.sub('',content_str)

print(content_string_cut)
```

```python
import jieba
seq_list = jieba.lcut(content_string_cut,cut_all=False)


print(seq_list)
```
